```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv('D:\SR14.csv')
df.head()
```

```python
df.rename(columns={'rad from body':'radfrombody'},inplace=True)
```

```python
df.rename(columns={'false teeth':'falteeth','using pacemaker':'pacemaker','thyroid scan':'thyroidscan'},inplace=True)
```

```python
x=df.iloc[:,0:10]
```

```python
from sklearn import preprocessing
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

```python
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
rs=[]
mae_list=[]
for i in range(49,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    rmc=RandomForestRegressor(n_estimators = 2, random_state = 34)
    rmc.fit(x_train,y_train)
    rmc_pred=rmc.predict(x_test)
    mae=metrics.mean_squared_error(rmc_pred,y_test)
    rs.append(i)
    mae_list.append(mae)
```

```python
df1_rmc=pd.DataFrame({'Random_state':rs,'Mean squared error':mae_list})
df1_rmc
```
